```sql
WITH date_homicides AS (
    SELECT CONCAT(MES_OCURR, '-', DIA_OCURR,'-' , ANIO_OCUR) AS DATE
    FROM dbo.DEFUN20
)

SELECT date_homicides.DATE,
ENT_OCURR AS ESTADO, 
COUNT(LISTA_MEX) AS HOMICIDIOS_MUJERES
FROM dbo.DEFUN22, date_homicides
WHERE LISTA_MEX = '55' AND SEXO = '2'
GROUP BY  ENT_OCURR, date_homicides.DATE
ORDER BY ENT_OCURR;
```

```sql
WITH date_homicides AS (
    SELECT
        CASE WHEN MES_OCURR = 99 THEN 1 ELSE MES_OCURR END AS MES,
        CASE WHEN DIA_OCURR = 99 THEN 1 ELSE DIA_OCURR END AS DIA,
        CASE WHEN ANIO_OCURR = 9999 THEN 2022 ELSE ANIO_OCURR END AS ANIO
    FROM dbo.DEFUN20
)

SELECT
    CONCAT(date_homicides.MES, '-', date_homicides.DIA, '-', date_homicides.ANIO) AS DATE,
    ENT_OCURR AS ESTADO, 
    COUNT(LISTA_MEX) AS HOMICIDIOS_MUJERES
FROM dbo.DEFUN22
JOIN date_homicides ON 1=1  -- JOIN sin condicin para combinar todas las filas
WHERE LISTA_MEX = '55' AND SEXO = '2'
GROUP BY ENT_OCURR, date_homicides.MES, date_homicides.DIA, date_homicides.ANIO
ORDER BY ENT_OCURR;
```

```sql
SELECT
    CONCAT(
        CASE WHEN MES_OCURR = 99 THEN 1 ELSE MES_OCURR END,
        '-',
        CASE WHEN DIA_OCURR = 99 THEN 1 ELSE DIA_OCURR END,
        '-',
        CASE WHEN ANIO_OCUR = 9999 THEN 2022 ELSE ANIO_OCUR END
    ) AS DATE, 
    COUNT(LISTA_MEX) AS HOMICIDIOS_MUJERES
FROM dbo.DEFUN22
WHERE LISTA_MEX = '55' AND SEXO = '2'
GROUP BY ANIO_OCUR, MES_OCURR, DIA_OCURR
ORDER BY DATE;
```
