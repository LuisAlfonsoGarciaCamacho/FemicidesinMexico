```python
import pandas as pd
from sqlalchemy import create_engine

# Configuración de la conexión a SQL Server LocalDB
database_url = 'mssql+pyodbc://(localdb)\\MSSQLLocalDB/defunciones?driver=ODBC+Driver+17+for+SQL+Server'
engine = create_engine(database_url)

# Lista de valores para la parte "NUM"
num_values = ["90", "91", "92", "93", "94", "95", "96", "97"] 

for num_value in num_values:
    # Consulta SQL
    sql_query = f"""
    SELECT
        CONCAT(
            CASE WHEN MES_OCURR = 99 THEN 1 ELSE MES_OCURR END,
            '-',
            CASE WHEN ANIO_OCUR = 9999 THEN 2022 ELSE ANIO_OCUR END
        ) AS DATE, 
        COUNT(LISTA_BAS) AS HOMICIDIOS_MUJERES
    FROM dbo.DEFUN{num_value}
    WHERE LISTA_BAS = '55' AND SEXO = '2'
    GROUP BY ANIO_OCUR, MES_OCURR
    ORDER BY DATE;
    """

    # Ejecutar la consulta y obtener el resultado en un DataFrame
    result_df = pd.read_sql(sql_query, engine)

    # Guardar los datos en un archivo CSV
    csv_filename = f"DEFUN_{num_value}.csv"
    result_df.to_csv(csv_filename, index=False)

    # Imprimir un mensaje de confirmación
    print(f"Resultados para dbo.DEFUN{num_value} guardados en {csv_filename}.")
    print("\n" + "="*40 + "\n")  # Separador visual
```

```python
import pandas as pd
import glob

# Carpeta donde se encuentran los archivos CSV
folder_path = '.'  # Puedes cambiar esto al directorio correcto

# Patrón para seleccionar archivos CSV
csv_pattern = 'DEFUN_*.csv'

# Obtener la lista de archivos CSV en la carpeta
csv_files = glob.glob(f"{folder_path}/{csv_pattern}")

# Verificar si hay archivos para procesar
if not csv_files:
    print("No se encontraron archivos CSV en la carpeta.")
else:
    # Inicializar un DataFrame vacío para contener los datos combinados
    combined_df = pd.DataFrame()

    # Iterar sobre los archivos CSV y concatenar al DataFrame
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        combined_df = pd.concat([combined_df, df], ignore_index=True)

    # Guardar el DataFrame combinado en un nuevo archivo CSV
    combined_csv_filename = 'Todas_defunciones.csv'
    combined_df.to_csv(combined_csv_filename, index=False)

    print(f"Archivos CSV combinados y guardados en {combined_csv_filename}.")
```
